* Aqui iremos abordar a metodologia principal para um projeto de dados **CRISP-DM**, **Extração de Dados**, **Analise Exploratória dos Dados**, **Modelagem** e **Avaliação**

# Metodologia CRISP-DM
* Cross Industry Standard Process of Data Mining

<img src= 'https://miro.medium.com/v2/resize:fit:640/format:webp/0*tA5OjppLK627FfFo' alt="Diagrama CRISP-DM">

1. **Business Understanding:**

A primeira etapa da metodologia consiste em identificar as reais necessidades do projeto.

    - Identificar o tema a ser abordado
    - Estudo e entendimento do(s) processo(s) dentro da empresa
    - Ouvir especialistas internos

Ao final desta etapa devemos ser capazes de responder:

    - Qual problema queremos resolver?
    - Por que ele é importante?
    - A partir de uma solução hipotética para o problema, como esta solução afetaria a empresa?
    - Temos capacidade de agir com base nessa solução ou também precisamos desenvolver essa habilidade?
    - Como vamos medir o sucesso da solução implementada (KPI)? O que faria dela um sucesso? Qual métrica será utilizada?

2. **Data Understanding:**

Esta etapa consiste em coletar, organizar e documentar todos os dados que se encontram disponíveis para realizar a análise exploratória. Se necessário utilizar diversas fontes de dados, precisará considerar como e quando integrá-las.

Perguntas a serem respondidas:

    - Os dados estão disponíveis?
    - Qual a dificuldade para extrair os dados? Essa ação é permitida? Temos acesso?
    - Definição da variável alvo (modelo supervisionado) ou não (modelo não supervisionado)

3. **Data Preparation:**

Nesta etapa, devemos tratar os dados para nos certificarmos que as informações estão de acordo com o esperado. Consistência de erros e ausência de valores deverão ser resolvidos para que possamos selecionar amostras aleatórias e utilizá-las para treino e testes.

Pontos importantes nesta etapa:

    - Limpeza dos dados: Tratar valores nulos e dados inconsistentes
    - Engenharia de Features: atributos derivados, atributos construídos a partir de um ou de muitos registros, em geral, tendem a vir de uma operação de agregação
    - Seleção de atributos relevantes para o modelo (feature selection)
    - Seleção da amostra

4. **Modelling:**

Nesta etapa definimos quais modelos testar. Cada projeto contém dados diferentes e terá um modelo ideal diferente, é muito importante que voce teste varios modelos, em conjunto a isso, cada objetivo de projeto esta sujeito a uma abordagem diferente sendo elas de regressão, classificação ou clusterização.

    - Técnica selecionada: Documente o modelo escolhido
    - Motivo da escolha: Alguns modelos possuem específicos pressupostos sobre os dados, por exemplo, todos os atributos possuem uma distribuição uniforme, não permitem valores nulos, atributos categóricos precisam ser transformados, etc
    - Avaliação do modelo: Descreva, de acordo com o modelo escolhido, um plano para treinar, testar e avaliar o modelo desenvolvido
    - Estimação de hiperparâmetros: Exemplo: GridSearchCV

*Obs: existem ferramentas para testar diversos modelos de vez, tanto modelos classicos quanto modelos caixa preta e ensembles.*


5. **Evaluating:**

Nessa fase, é preciso avaliar os resultados e levantar todas as possibilidades de variações que os dados possam ter, analisando se existe algum fator que tenha sido negligenciado e até que ponto o modelo atende aos objetivos do negócio. Caso o modelo não esteja performando como esperado, devemos retornar a primeira etapa para entender o negócio e os dados (assim como mostra o diagrama).

    - Note que assim como cada subclasse de modelo tem seus objetivos diferentes (regressão, classificação ou clusterização) eles também tem metricas de avaliação diferentes.

6. **Deployment:**

Nesta etapa, o modelo é enviado para produção e implementado em protótipo. Será necessário acompanhar a performance do modelo em um período alinhado com a área de negócio. Realizar estratégia de rollout.

Em particular, cada projeto possui seu tempo de duração e deve ser planejado de acordo com os recursos disponíveis.

Pipelines de Avaliação do modelo são muito importantes pra manter o modelo estavel a partir do momento que ele começa a trabalhar com dados reais (dados não vistos durante a fase de treinamento), a porcentagem de acerto/erro e resultados das avaliações da fase anterior devem sempre se comportar da mesma maneira, caso contrario acontece o fenomeno Data Drift
